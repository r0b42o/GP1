$ m = \frac{ΔF_g}{Δg}$

Für die Masse des IOLab-Geräts sah diese Rechnung beispielhaft folgend aus:

$ΔF_g = (-1,97±0,02)N$

$Δg = (-9,80±0,14)\frac{m}{s^2}$

$ m = \frac{ΔF_g}{Δg} = \frac{(-1,97±0,02)N}{(-9,80±0,14)\frac{m}{s^2}} =(0,201±0,004)kg$

Beispielsweise für die Schwingung des IOLab-Geräts:

$Δt_{Intervall} = (2,99±0,03)s$

$T = \frac{Δt_{Intervall}}{4} = \frac{(2,99±0,03)s}{4} = (0.748±0.007)s$

```python
def linF(x, a,b):
    return a*x + b

new_m1 = ufloat(2.23,0.02)
new_m2 = ufloat(1.817,0.015)
new_m3 = ufloat(1.648,0.018)

w1 = ufloat(8.40,0.08)
w2 = ufloat(6.61,0.05)
w3 = ufloat(6.22,0.04)


Mass = [new_m1.nominal_value, new_m2.nominal_value, new_m3.nominal_value]
Winkelf = [w1.nominal_value, w2.nominal_value, w3.nominal_value]

error_m1=new_m1.std_dev
error_m2=new_m2.std_dev
error_m3=new_m3.std_dev
error_w1=w1.std_dev
error_w2=w2.std_dev
error_w3=w3.std_dev
m_errorbars = [error_m1, error_m2, error_m3]
w_errorbars = [error_w1, error_w2, error_w3]

fig, axs = plt.subplots(1,1,figsize=(13, 10));
plt.scatter(Mass, Winkelf, label = 'Daten', color='green', s=20, zorder= 5)
axs.errorbar(Mass,Winkelf,yerr = w_errorbars,linestyle='none',capsize = 4, label = r'$δω$');
axs.errorbar(Mass,Winkelf,xerr = m_errorbars,linestyle='none',capsize = 4, label = r'$δm$');
sc.optimize.curve_fit(linF, Mass, Winkelf, absolute_sigma=True)
popt, pcov = sc.optimize.curve_fit(linF, Mass, Winkelf)
a, b = popt
print('Geradengleichung:')
print('y = %.5f * x + %.5f \n' % (a, b))

x_line = arange(0, max(Winkelf), 1)
y_line = linF(x_line, a, b)
perr = np.diag(pcov)
a1 = ufloat(a, perr[0])
b1 = ufloat(b, perr[1])
plt.plot(x_line, y_line, '--', color='red', label = r'y(x) = $3.86(21)\;\sqrt{\frac{1}{kg\cdot s^2}} - 0.2(8)\;\frac{1}{s}$ ')

print('a =', a1)
print('b =', b1)
k = a1**2
print('Federkonstante k =', k, 'N/m')
print('\n')




axs.set_xlabel(r'Masse $\hat{m}$ in $\sqrt{\frac{1}{kg}}$', fontsize = 20);
axs.set_ylabel(r'Winkelfrequenz $\omega$ in $\frac{1}{s}$', fontsize = 20);

axs.legend(loc = 'lower right',fontsize=15)

plt.axis([1.6, 2.3, 6, 8.5])
plt.title('Lineare Regression zur Bestimmung der Federkonstanten k', fontsize = 20)
plt.grid()

plt.show()
```
